The purpose of this code is to save still images from traffic cameras. These images are hosted online as a static .jpg url, so you just need to save the same .jpg url to a local image using a different filename each time. Images will be collected for an indefinite period of time.

```python
import urllib.request
import time
from datetime import datetime, timedelta
i=0
while True:
    now = datetime.now() - timedelta(hours=1, minutes=0)
    dt_string = now.strftime("%Y%m%dT%H%M%S")
    print("date and time =", dt_string)
    
    urllib.request.urlretrieve("https://www.houstontranstar.org/snapshots/cctv/166.jpg", "166_" + dt_string + ".jpg")
    print("166_" + dt_string+".jpg")
    time.sleep(60)
```
